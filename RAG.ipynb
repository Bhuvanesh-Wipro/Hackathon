```python
import textwrap

def wrap_text_preserve_newlines(text, width=110):
    # Split the input text into lines based on newline characters
    lines = text.split('\n')

    # Wrap each line individually
    wrapped_lines = [textwrap.fill(line, width=width) for line in lines]

    # Join the wrapped lines back together using newline characters
    wrapped_text = '\n'.join(wrapped_lines)

    return wrapped_text
```

```python
print(wrap_text_preserve_newlines(str(documents[15].page_content)))
```

```python
# Text Splitter
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1500, chunk_overlap=100)
docs = text_splitter.split_documents(documents)
```

```python
# Embeddings
from langchain.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings()
```

```python
# Vectorstore: https://python.langchain.com/en/latest/modules/indexes/vectorstores.html
from langchain.vectorstores import Chroma

db = Chroma.from_documents(docs, embeddings)
```

```python
print(str(docs[0].page_content))
```

```python
#QA Chain
from langchain.chains.question_answering import load_qa_chain
from langchain import HuggingFaceHub
```

```python
llm=HuggingFaceHub(repo_id="HuggingFaceH4/zephyr-7b-beta", model_kwargs={"temperature":0.02, "max_length":512})
```

```python
chain = load_qa_chain(llm, chain_type="stuff")
```

```python
query = "what are all  established in the year 1998"
search_type="mmr"
docs = db.search(query,search_type)
chain.run(input_documents=docs, question=query)
```
